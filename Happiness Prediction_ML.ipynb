```python
q1 = x.quantile(0.25)
q3 = x.quantile(0.75)
iqr = q3 - q1
```

```python
std_err_high = (q3.Standard_Error + (1.5 * iqr.Standard_Error))
print(std_err_high)
index = np.where(x['Standard_Error'] > std_err_high) 
index
x = x.drop(x.index[index])
x.shape
y = y.drop(y.index[index])
y.shape
```

```python
std_err_low = (q1.Standard_Error - (1.5 * iqr.Standard_Error))
print(std_err_low)
index = np.where(x['Standard_Error'] < std_err_low) 
index
x = x.drop(x.index[index])
x.shape
y = y.drop(y.index[index])
y.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
y
```

```python
x
```

```python
maxAccu = 0
maxRS = 0
for i in range(0,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20, random_state =i)
    LR.fit(x_train,y_train)
    predrf = LR.predict(x_test)
    acc = accuracy_score(y_test,predrf)
    print('Accuracy:',acc,' Random_state:',i)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
        print('Accuracy:',maxAccu,'Random_state:',i)
```
