```python
# deleting all columns which are insignificant to making prediction model
df = df.drop(['MinTemp','MaxTemp','Date'], axis=1)
```

```python
df
```

```python
# filling null values in dependants column by using 'ffill' method
df['Evaporation'].fillna(method ='ffill', inplace = True)
```

```python
df['Sunshine'].fillna(method ='ffill', inplace = True)
```

```python
df['Cloud3pm'].fillna(method ='ffill', inplace = True)
```

```python
df['Cloud9am'].fillna(method ='ffill', inplace = True)
```

```python
# deleting records with null values as columns with most null values are now okay
df = df.dropna()
df = df.reset_index()
```

```python
df = df.drop(['index'], axis=1)
```

```python
df
```

```python
# encoding string type data to use them in plotting and making model on basis of observations
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['Location'] = LE.fit_transform(df['Location'])
```

```python
LE2 = LabelEncoder()
df['WindGustDir'] = LE2.fit_transform(df['WindGustDir'])
```

```python
LE3 = LabelEncoder()
df['WindDir9am'] = LE3.fit_transform(df['WindDir9am'])
```

```python
LE4 = LabelEncoder()
df['WindDir3pm'] = LE4.fit_transform(df['WindDir3pm'])
```

```python
LE5 = LabelEncoder()
df['RainToday'] = LE5.fit_transform(df['RainToday'])
```

```python
LE6 = LabelEncoder()
df['RainTomorrow'] = LE6.fit_transform(df['RainTomorrow'])
```

```python
# checking corelation
corr_mat = df.corr()
plt.figure(figsize=(22,7))
sns.heatmap(corr_mat,annot=True,linewidths=0.1,linecolor='black',fmt='0.2f')
```

```python
# plotting density of values in each column
import warnings
plt.figure(figsize=(15,20),facecolor='white')
plotnumber = 1

for column in df:
    if plotnumber<=19:
        ax = plt.subplot(5,4,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# Preprocessing
# spliting dependent and independent variables on x and y axis
x = df.drop('RainToday',axis=1)
y = df['RainToday']
```

```python
x
```

```python
y
```

```python
x.dtypes
```

```python
# making scatter plot of Income vs other independent variables in dataset for observations
plt.figure(figsize=(15,20),facecolor='white')
plotnumber = 1

for column in x:
    if plotnumber<=10:
        ax = plt.subplot(4,4,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=10)
        plt.ylabel('Rain_Tomorrow',fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# use Random Forest to make model for Income prediction
from sklearn.model_selection import train_test_split
import seaborn as sns
import pickle

import warnings
warnings.filterwarnings('ignore')

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=60)
y_train.head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators= 101, n_jobs=-1)
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
# find accuracy of predictions made by comparing to test sample
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# save model in pickle file
import pickle
filename= 'Rainfall_Prediction_model.pickle'
pickle.dump(RandomForestClassifier,open(filename,'wb'))
```

```python
# now for rainfall value prediction we will make separate dataset split
x1 = df.drop('Rainfall',axis=1)
y1 = df['Rainfall']
```

```python
x1
```

```python
y1
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sns
import pickle

import warnings
warnings.filterwarnings('ignore')

x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.25,random_state=70)
y1_train.head()
```

```python
# using Linear Regression to train model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import pickle

import warnings
warnings.filterwarnings('ignore')
l_reg = LinearRegression()
l_reg.fit(x1_train,y1_train)
```

```python
y1_pred = l_reg.predict(x1_test)
```

```python
y1_pred
```

```python
plt.scatter(y1_test,y1_pred)
plt.xlabel('Actual Status')
plt.ylabel('Predicted Status')
plt.title('Rainfall Actual vs Prediction')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_absolute_error(y1_test,y1_pred)
```

```python
#saving model
import pickle
filename= 'rainfall_model.pickle'
pickle.dump(l_reg,open(filename,'wb'))
```
